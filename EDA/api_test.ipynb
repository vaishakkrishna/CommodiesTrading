```python
import requests
import pandas as pd
from time import sleep
```

```python
with open('symbols.txt', 'r') as f:
    symbol_list = f.readlines()
    symbol_list = ["%40" + x.strip() + "%7C" for x in symbol_list]
    symbols = "".join(symbol_list)
    symbols = symbols[:-3]
    print(symbols)

s = f"https://quote.cnbc.com/quote-html-webservice/restQuote/symbolType/symbol?symbols={symbols}&requestMethod=itv&noform=1&partnerId=2&fund=1&exthrs=1&output=json&events=1"
x = requests.get(s)
j = x.json()
j= j['FormattedQuoteResult']['FormattedQuote']
sleep(0.1)
    
```
